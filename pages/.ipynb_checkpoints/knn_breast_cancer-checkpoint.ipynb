```python

import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data .csv')
df.head()
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
df.head()
```

```python
df["diagnosis"].unique()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,1:], df.iloc[:,0],test_size=0.2,random_state=2)  #train_test_split(x_dataset, y_dataset,test_size=0.2)
# test 20% data 20/100  569x0.2= 113.8=114
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
X_train
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=4)
```

```python
knn.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = knn.predict(X_test)

accuracy_score(y_test, y_pred)
#97.36%  108 predictions are right
```

```python
y_pred
```

```python
y_test.values
```

```python
scores = []

for i in range(1,16):

    knn = KNeighborsClassifier(n_neighbors=i)

    knn.fit(X_train,y_train)

    y_pred = knn.predict(X_test)

    scores.append(accuracy_score(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt

plt.plot(range(1,16),scores)
```
